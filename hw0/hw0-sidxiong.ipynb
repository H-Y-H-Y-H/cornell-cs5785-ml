```python
# _ = np.dtype([('f1', np.float), ('f2', np.float), ('f3', np.float), ('f4', np.float),
#              ('label', 'S10')])
data = np.loadtxt("iris.data", delimiter=',',dtype=str)
# print data[:10]

label_dict = {l:i for i, l in enumerate(set(data[:, -1]))}
# print label_dict
```

```python
import itertools

colors = ['r', 'g', 'b']
NUM = 4

plt.figure(figsize=(8, 5))
for i, idx in enumerate(list(itertools.combinations(range(NUM), 2))):
    plt.subplot(2, 3, i+1)
    plt.scatter(data[:, idx[0]], data[:, idx[1]],
               c=[colors[label_dict[x]] for x in data[:, -1]])

plt.tight_layout()
```
